```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [11, 7]
df = pd.read_csv('datasets/office_episodes.csv')
df.head()
```

```python
#Creating a list of colors that shows different colors based on the scaled ratings
color = []
for index, rating in df.iterrows():
    if rating['scaled_ratings'] < 0.25:
        color.append("red")
    elif rating['scaled_ratings'] >= 0.25 and rating['scaled_ratings'] < 0.50:
        color.append("orange")
    elif rating['scaled_ratings'] >= 0.50 and rating['scaled_ratings'] < 0.75: 
        color.append("lightgreen")
    else: 
        color.append("darkgreen")
        
# Creating a list showing the episode sizes that are the largest with a guest appearance        
sizes = []
for index, size in df.iterrows(): 
    if size['has_guests'] == True:
        sizes.append(250)
    else: 
        sizes.append(25)
```

```python
# Import matplotlib.pyplot under its usual alias and create a figure
import matplotlib.pyplot as plt

fig = plt.figure()

# Creating a scatter plot of The Office episodes, that show the popularity and quality of the series varied over time
plt.scatter(df['episode_number'], df['viewership_mil'], c=color, s=sizes)


plt.title("Popularity, Quality, and Guest Appearances on the Office")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")

plt.style.use('ggplot')
plt.show()
```

```python
#showing the name of the guest stars that were in the most-watched Office episode
df.nlargest(1, 'viewership_mil')['guest_stars']
top_star = 'Cloris Leachman'
```
